```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

# 1. Pie chart


A pie chart is used to visualize the composition ratio of categorical data.


use the pie() function for this. 

```python
programs = ['Python', 'R', 'Java']
use = [96, 53, 32]
plt.pie([96, 53, 32], labels = ['Python', 'R', 'Java'], colors=['red','yellow','blue'])
plt.title('programming language', fontsize=15)  #fontsize는 글자크기를 조정할 수 있는 인자입니다!
plt.show()
```

```python
plt.pie(use, labels = programs, colors=['red','yellow','blue'],
       autopct='%1.2f%%', shadow=True, startangle=90,
        textprops={'fontsize':12})
plt.show()
```

```python
plt.pie(use, labels = programs, colors=['red','yellow','blue'], autopct='%1.2f%%', shadow=True, startangle=90, textprops={'fontsize':12})
#autopct='%1.2f%%' : float
#startangle=90 : The graph starts from left
#textprops={'fontsize':12} : use this to change the font size of the text in the pie chart
plt.title('programming language', fontsize=15) 
plt.show()
```

# 2. Bar plot

A bar graph, like a pie chart, is used for categorical data.


You can draw a bar graph using bar() and barh() functions.


The main parameters include color, alpha, and width.

use built-in datasets from the seaborn library as data for our graphs.

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

Take a quick look at the structure of the data

```python
tips.head()
```

The “day” column in the tips dataset is a categorical column.

After calculating the total tip amount for each category of “day.” Based on this, draw a bar plot.


```python
day_sum = tips.groupby('day').tip.sum()
day_sum
```

```python
day_sum
```

```python
plt.bar(['Thur', 'Fri', 'Sat', 'Sun'], day_sum)
plt.title('Total tips', fontsize=12)
plt.xlabel('days')
plt.ylabel('tips')
plt.show()
```

```python
plt.barh(['Thur', 'Fri', 'Sat', 'Sun'], day_sum,
       color='turquoise', alpha=0.8)
plt.title('Total tips', fontsize=12)
plt.xlabel('days')
plt.ylabel('tips')
plt.show()
```

The barh() function is used to draw a horizontal bar graph.
It works the same way as bar(), but since the values for the x-axis are displayed on the y-axis, need to be careful when setting the axes and related options.

```python
plt.barh(['Thur', 'Fri', 'Sat', 'Sun'], day_sum, color='hotpink', alpha=0.8, height=0.5)
# plt.title('Total tips', fontsize=12)
# plt.ylabel('days')
# plt.xlabel('tips')
# plt.show()
```

Draw grouped bar graphs to compare how tips change by day of the week, according to the time.

```python
tips.groupby('time').tip.sum()
```

there are both lunch time and dinner time.

```python
lunch_tips = tips[tips['time']=='Lunch'].groupby('day').tip.sum()
dinner_tips = tips[tips['time']=='Dinner'].groupby('day').tip.sum()

print(lunch_tips)
print(dinner_tips)
```

To draw two bar graphs side by side, think of it as shifting the x-values of one of the graphs.

```python
x=np.arange(4)
plt.bar(x, lunch_tips, color='yellow', alpha=0.5, width=0.4, label='lunch')
plt.bar(x+0.5, dinner_tips, color='blue', alpha=0.5, width=0.4, label='dinner') #0.5만큼 가로축 값에 더해줌으로써 그래프를 오른쪽으로 이동했다고 생각하면 됩니다!
plt.title('Total tips', fontsize=12)
plt.xticks(x, labels=['Thur', 'Fri', 'Sat', 'Sun'])
plt.legend()
plt.xlabel('days')
plt.ylabel('tips')
plt.show()
```

# 3. Histogram


It is commonly used to plot probability or frequency distributions, and you can use the hist() function.

The main parameters include bins, density, color, and alpha.

bins: Number of bars (bins)

density: Whether to display as a probability density

color: Color of the graph

alpha: Transparency





draw a graph of the standard normal distribution.

```python
data = np.random.randn(1000)
```

```python
plt.hist(data, bins=30, density=True)
plt.show()
```

Draw overlaying two graphs and two normal distribution graphs with different means and variances.

```python
x = np.random.normal(5, 5, 5000)
y = np.random.normal(0, 1, 5000)
```

```python
plt.hist(x, bins=30, density=True, color='red', alpha=0.7, label='N(5,25)')
plt.hist(y, bins=30, density=True, color='blue', alpha=0.7, label='N(0,1)')
plt.legend()
plt.show()
```

By writing and displaying the code for the two graphs one after another like this,they are plotted overlapping each other.

# 4. Scatter plot

A scatter plot is usually used to examine the relationship between two variables.

You can use the scatter() function, and the main parameters include marker, c, and s.

marker: Marker shape

c: Marker color

s: Marker size

alpha: Transparency

Note that for scatter plots, the color is set using c instead of color—which is different from other plot types.

Use the built-in iris dataset from seaborn to create a scatter plot.

The iris dataset contains measurements of sepal (width and length) and petal (width and length) for different species of iris flowers.


```python
iris = sns.load_dataset('iris')
iris.head()
```

 The length and width of the sepals and petals vary by species and individual flower.
 Visualize their relationship with a scatter plot.

```python
plt.scatter(iris['petal_length'], iris['petal_width'], s=200, c='gold',  alpha=0.3)
plt.title('length and width of petal',  fontsize=15)
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

petal width and length are proportional to each other.

Adjust the parameters we mentioned earlier.

The marker size to 200, the color to gold, and adjusted the transparency.


```python
plt.scatter(iris['petal_length'], iris['petal_width'], s=200, c='gold', alpha=0.3)
plt.title('length and width of petal', fontsize=15)
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

Adjust the size and color of the points based on other variable values.
Examine the relationship between sepal length and sepal width.
Make the petal width proportional to the marker size, and represent petal length using color variations in the graph.

```python
plt.scatter(iris['sepal_length'], iris['sepal_width'])
```

```python
plt.scatter(iris['sepal_length'], iris['sepal_width'], 
           s=iris['petal_length']*100, c=iris['petal_width'], alpha=0.3)
plt.title('length and width of sepal', fontsize=15)
plt.xlabel('length')
plt.ylabel('width')
plt.colorbar()  #이 코드를 추가하면 color bar 가 생성되어 점의 색깔이 어떤 값을 의미하는지 알 수 있습니다.
```

```python
plt.scatter(iris['sepal_length'], iris['sepal_width'], s=iris['petal_length']*100, c=iris['petal_width'], alpha=0.3)
plt.title('length and width of sepal', fontsize=15)
plt.xlabel('length')
plt.ylabel('width')
plt.colorbar()  #By adding this code, a color bar will be created, which helps indicate what values the colors of the points represent.
plt.show()
```
