```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

plt.figure() is a function used to specify the size of a blank canvas. You can set the exact size using the figsize parameter.

plt.show() is used to display the figure.

plt.figure(figsize=(m,n)) : Set the canvas size
plt.show() : Display the figure

To draw a line graph, you can apply an array or list containing the X and Y coordinate values to the plot() function. 
If you only provide one set of values, it will be recognized as Y values, and the X coordinates will automatically be set as [0, 1, 2, 3, ...]. 

```python
plt.figure(figsize=(5,3))
plt.plot([3,4],[6,8])
plt.show()
```

Change the title and axes.
Below are functions used to modify the title and axes.

title() : Function to add a graph title

xlabel() : Function to add an x-axis label

ylabel() : Function to add a y-axis label

xticks() : Function to set x-axis tick marks (can be specified with labels)

yticks() : Function to set y-axis tick marks (can be specified with labels)

axis() : Function to set the range of the x and y axes

```python
x=[1,2,3,4]
y=[2,4,6,8]

plt.figure(figsize=(5,3))

plt.plot(x, y)
plt.title('line plot example')
plt.xlabel('x label')
plt.ylabel('y label')
plt.xticks(x, labels=['a','b','c','d'])  #x축 눈금 표시
plt.yticks(y, labels=[2,4,6,8])  #y축 눈금 표시
plt.axis([0,5,0,10])  #x, y축 범위 지정([xmin, xmax, ymin, ymax]의 형태로 넣어주면 됨!)
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]

plt.figure(figsize=(5,3))

plt.plot(x, y)


plt.title('line plot example')  #제목 추가
plt.xlabel('x label')  #x축 레이블 추가
plt.ylabel('y label')  #y축 레이블 추가
plt.xticks(x, labels=['a','b','c','d'])  #x축 눈금 표시
plt.yticks(y, labels=[2,4,6,8])  #y축 눈금 표시
plt.axis([0,5,0,10])  #x, y축 범위 지정([xmin, xmax, ymin, ymax]의 형태로 넣어주면 됨!)
plt.show()
```

Set the line style, line color, and marker within plot() by using various parameters.


Line Types 

|code|type|
|:---:|:---:|
|'-'|solid line|
|'--'|dashed line|
|'-.'|dash-dot line|
|':'|dotted line|


색상

|code|type|
|---|:---:|
|'b' or 'blue'|blue|
|'g' or 'green'|green|
|'y' or 'yellow'|yellow|
|'k' or 'black'|black|
|'w' or 'white'|white|


마커

|code|type|
|:---:|:---:|
|'.'|point marker|
|'o'|circle marker|
|'v'|triangle marker|
|'s'|square marker|
|'*'|star marker|
|'+'|plus marker|

```python
plt.plot(figsize=(5,3))
plt.plot([1,5,8,15], linestyle=':', color='red', marker='o')
plt.title('line plot example')
plt.show()
```

Specify the 'color-marker-line style' all together without using separate parameters within the function.

```python
plt.plot(figsize=(5,3))
plt.plot([1,5,8,15], 'ro:') #색상 red, 원 모양 마커, 점선 적용
plt.title('line plot example')
plt.show()
```

Add a legend this time.

Add a legend using the legend() function. After specifying the label parameter within the plot function, it will recognize it as a legend.

```python
x1 = np.random.randint(1,10,10)
```

```python
x2 = np.random.randint(1,10,10)
```

```python
plt.plot(x1, label='x1')
plt.plot(x2, label='x2')
plt.title('random numbers(from 1 to 10)')
plt.legend()
plt.show()
```

Set the position of the legend using the loc parameter.

Additionally, specify the number of columns in the legend using the ncol parameter.

```python
x1=np.random.randint(1,10,10)
x2=np.random.randint(1,10,10)
plt.plot(x1, label='x1')
plt.plot(x2, label='x2')
plt.title('random numbers(from 1 to 10)')
plt.legend(loc=(0.5, 0.7), ncol=2)  #loc : x축, y축을 1이라고 할 때, 좌표로 위치를 표현한다고 생각!, ncol : 디폴트는 1이다!
plt.show()
```
